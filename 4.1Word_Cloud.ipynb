```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
import pandas as pd

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches # needed for waffle Charts
```

```python
import urllib

# open the file and read it into a variable alice_novel
yesi_cv= urllib.request.urlopen('https://docs.google.com/document/d/1IIQSQXVrBmsNtTOFAs7viLhGGUB38R8I/edit?usp=sharing&ouid=108960027149883171754&rtpof=true&sd=true').read().decode("utf-8")
```

```python
#Next, let's use the stopwords that we imported from `word_cloud`. 
#We use the function *set* to remove any redundant stopwords.
stopwords = set(STOPWORDS)
```

```python
#Create a word cloud object and generate a word cloud. For simplicity, let's generate a word cloud using only the first 20 words in the cv.
# instantiate a word cloud object
yesi_cv_wc = WordCloud(
    background_color='white',
    max_words=5,
    colormap='Accent',
    stopwords=stopwords
)

# generate the word cloud
yesi_cv_wc.generate(yesi_cv)
```

```python
# display the word cloud
plt.imshow(yesi_cv_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
